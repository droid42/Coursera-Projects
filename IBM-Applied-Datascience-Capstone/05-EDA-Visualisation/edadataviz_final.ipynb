First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass is also important; it seems the more massive the payload, the less likely the first stage will return.


```python
# Define a custom color palette
custom_palette = {
    0: 'red',
    1: 'green'
}
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", palette=custom_palette, data=df, aspect = 3)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

We see that different launch sites have different success rates.  <code>CCAFS LC-40</code>, has a success rate of 60 %, while  <code>KSC LC-39A</code> and <code>VAFB SLC 4E</code> has a success rate of 77%.


Next, let's drill down to each site visualize its detailed launch records.

```python
# find success rate for flightnumber > 62
df[df['FlightNumber'] > 62]['Class'].mean()
```

```python
df['PayloadCategory'] = pd.cut(df['PayloadMass'], bins=[0, 4000, 8000, 10000, 20000], labels=['0-4000', '4000-8000', '8000-10000', '10000-20000'])

# plot two charts side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# plot success rate for flightnumber > 62, grouped by payload mass categories

# plot side by side Flight Number <=62
df[df['FlightNumber'] <= 62].groupby('PayloadCategory')['Class'].mean().plot(kind='bar', ax=axs[0])
axs[0].set_title('Success Rate for Flight-Number <= 62, grouped by Payload Mass Categories')
axs[0].set_xlabel('Payload Mass Category')
axs[0].set_ylabel('Success Rate')

# plot success rate for flightnumber > 62, grouped by payload mass categories
df[df['FlightNumber'] > 62].groupby('PayloadCategory')['Class'].mean().plot(kind='bar', ax=axs[1])
axs[1].set_title('Success Rate for Flight-Number > 62, grouped by Payload Mass Categories')
axs[1].set_xlabel('Payload Mass Category')
axs[1].set_ylabel('Success Rate')

# Standardize y-axis range across all subplots
for ax in axs:
    ax.set_ylim(0, 1)

plt.tight_layout()
plt.show()

```

```python
### TASK  3: Visualize the relationship between success rate of each orbit type
# Plotting with catplot
custom_palette = {
    0: 'red',
    1: 'green'
}
sns.catplot(data=df, x='LaunchSite', y='Orbit', hue='Class', kind='strip', height=6, aspect=1.5, palette=custom_palette)
plt.title('Scatter Plot of Launch Site vs Orbit')
plt.show()
```

Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
df_oc = df.groupby('Orbit')['Class'].value_counts().unstack(fill_value=0)
labels = ['failure', 'success']
df_oc.plot(kind='bar', stacked=False, colormap='RdYlGn')
plt.title('Bar Chart of Orbit vs Class')
plt.xlabel('Orbit')
plt.ylabel('Count')
plt.legend(title='Class', labels=labels)
plt.show()
```

```python
### TASK  6: Visualize the launch success yearly trend
```

You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
df.groupby('Date')['Class'].mean()
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
sr_df = df.groupby('Date')['Class'].mean().reset_index()
sr_df.head()

plt.figure(figsize=(10, 6))
sns.lineplot(data=sr_df, x='Date', y='Class', marker='o')
plt.title('Yearly Success Rate')
plt.xlabel('Year')
plt.ylabel('Success Rate')
plt.ylim(0, 1)  # Since success rate is a proportion
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020


```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
# add trendline
sr_df['Date_numeric'] = sr_df['Date'].astype(int)

sr_df = sr_df.groupby('Date_numeric')['Class'].mean().reset_index()
sr_df.head()

plt.figure(figsize=(10, 6))
sns.lineplot(data=sr_df, x='Date_numeric', y='Class', marker='o')

# Add a red trendline
# Convert 'Date' to numeric because regplot requires numeric x values
sns.regplot(data=sr_df, x='Date_numeric', y='Class', scatter=False, color='red', label='Trendline')

plt.title('Yearly Success Rate')
plt.xlabel('Year')
plt.ylabel('Success Rate')
plt.ylim(0, 1)  # Since success rate is a proportion
plt.show()
```
